# 03 — Modeling

Train and compare multiple classifiers.

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np

from config import DATA_PROCESSED
from src.models import get_models, save_model
from src.evaluation import evaluate, plot_confusion_matrix, cross_validate_models
```

## Load Processed Data

```python
X_train = pd.read_csv(DATA_PROCESSED / 'X_train.csv')
X_test = pd.read_csv(DATA_PROCESSED / 'X_test.csv')
y_train = pd.read_csv(DATA_PROCESSED / 'y_train.csv').squeeze()
y_test = pd.read_csv(DATA_PROCESSED / 'y_test.csv').squeeze()
print(f'Train: {X_train.shape}, Test: {X_test.shape}')
```

## Cross-Validation Comparison

```python
models = get_models()
cv_results = cross_validate_models(models, X_train, y_train)
cv_results.sort_values('mean_accuracy', ascending=False)
```

## Train and Evaluate Each Model

```python
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    metrics = evaluate(model, X_test, y_test, model_name=name)
    results.append(metrics)
    save_model(model, name)
```

```python
results_df = pd.DataFrame(results).sort_values('accuracy', ascending=False)
results_df
```

## Confusion Matrices

```python
for name, model in models.items():
    plot_confusion_matrix(model, X_test, y_test, model_name=name, save=True)
```
