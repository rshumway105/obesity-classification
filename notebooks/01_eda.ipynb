# 01 — Exploratory Data Analysis

Explore distributions, correlations, and class balance.

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from config import TARGET, TARGET_CLASSES, CATEGORICAL_FEATURES, CONTINUOUS_FEATURES
from src.data_loader import load_raw, validate

%matplotlib inline
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## Load Data

```python
df = load_raw()
validate(df)
df.head()
```

## Target Distribution

```python
order = TARGET_CLASSES
ax = sns.countplot(data=df, y=TARGET, order=order, palette='viridis')
ax.set_title('Class Distribution')
plt.tight_layout()
plt.show()
```

## Continuous Feature Distributions

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for ax, col in zip(axes.ravel(), CONTINUOUS_FEATURES):
    df[col].hist(ax=ax, bins=30, edgecolor='black')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

## Categorical Feature Distributions

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for ax, col in zip(axes.ravel(), CATEGORICAL_FEATURES):
    df[col].value_counts().plot.bar(ax=ax, edgecolor='black')
    ax.set_title(col)
    ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

## Correlation Heatmap (Continuous Features)

```python
corr = df[CONTINUOUS_FEATURES].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

## Pairplot: Height, Weight, Age by Obesity Level

```python
sns.pairplot(
    df, vars=['Height', 'Weight', 'Age'],
    hue=TARGET, hue_order=TARGET_CLASSES,
    palette='viridis', diag_kind='kde', plot_kws={'alpha': 0.5}
)
plt.suptitle('Pairplot', y=1.02)
plt.show()
```
